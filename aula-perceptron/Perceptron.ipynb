```python
#Importando a Biblioteca Pandas para a leitura de Bancos de Dados
import pandas as pd

#Lendo, criando as entradas e saída e armazenando os valores. 
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
#Preparado o dataset em dados para Treinamento e Validação
from sklearn.model_selection import train_test_split
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25)
```

```python
#Importando as Bilbiotecas do Keras para a Criação da Rede Perceptrn
import keras
from keras.models import Sequential
from keras.layers import Dense

#Montando a Rede
classificador = Sequential()
classificador.add(Dense(units=1, activation='sigmoid', kernel_initializer = 'random_uniform', input_dim = 30))


#classificador.add(Dense(units=16, activation='relu', kernel_initializer = 'random_uniform', input_dim = 30))
#classificador.add(Dense(units = 1, activation='sigmoid'))
```

```python
classificador.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
classificador.fit(previsores_treinamento, classe_treinamento, batch_size=10, epochs = 100)
```

```python
previsoes = classificador.predict(previsores_teste)
previsoes = (previsoes > 0.5)
#print(previsoes)
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
acuracia = accuracy_score(classe_teste, previsoes)
precisao = precision_score(classe_teste, previsoes)
recall = recall_score(classe_teste, previsoes)
f1score = f1_score(classe_teste, previsoes)
matriz = confusion_matrix(classe_teste, previsoes)
```

```python
#Imprimir os valores
print('Acurácia: %f' % acuracia)
print('Precision: %f' % precisao)
print('Recall: %f' % recall)
print('F1-Score: %f' % f1score)
print(matriz)
```
